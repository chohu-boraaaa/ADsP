## 선 그래프와 산점도 확인

### 산점도
* 두 개의 변수 혹은 여러 개의 변수 간 상관관계 확인
* 수치로 된 자료에 대해 두 변수 간 관계 산점도로 확인
* 다변량 자료는 pairs() 함수를 사용 -> 여러 변수 들 간 관계 한눈에 파악 가능

```r
# 두 개의 변수 상관관계 테스트
# x1 : 학교까지의 거리
# x2 : 학생들의 성적
x1 = c(10, 1, 5, 2, 4, 6, 1, 3, 8)
x2 = c(82, 94, 88, 90, 87, 84, 91, 87, 83)
```

```r
cor(x1, x2)
```

```r
plot(x1, x2, main="correlation test")
abline(lm(x2~x1), col="red", lwd=2, lty=1)
```

```r
pairs(iris, col=iris$Species)
```

### 선그래프
* 두 개의 변수에서 하나의 변수가 시간변수일 때 많이 사용
* 시계열 분석을 위해 자주 사용되는 그래프

```r
# late student analysis
month = 1:12
late = c(4, 5, 7, 10, 12, 13, 6, 5, 4, 2, 8, 9)
```

```r
plot(month, late, main="late student", type="l", lwd=2, xlab="month", ylab="count")
```

```r
# 점그래프 그리기
plot(month, late, main="late student", type="b", lwd=2, col="red", xlab="month", ylab="count")
```

```r
# 계단형 그래프 그리기
plot(month, late, main="late student", type="s", lwd=2, col="blue", xlab="month", ylab="count")
```

```r
# 여러 개의 선 그래프 그리기
month=1:12
late2 = c(8, 7, 5, 3, 11, 7, 8, 3, 5, 10, 12, 9)
plot(month, late, main="late student", type="l", lwd=2, col="pink", xlab="month", ylab="count")
lines(month, late2, type="l", col="sky blue")
```
