## 9. 신경망

### 1. 신경망
- 1940년대 중반 임계논리 알고리즘을 바탕으로 신경망을 위해 제안된 수학모델
- 인간의 학습 기능을 모방하는 방법
- 신경망은 인간의 뇌가 패턴을 인식하는 방식을 모사한 알고리즘으로 이미지, 소리, 문자, 시계열 데이터 등의 패턴을 인식할 수 있다.
- 신경망은 마치 로지스틱 회귀분석 혹은 다중 회귀분석을 여러 개 연결한 것으로 데이터를 분류하거나 예측할 때 모두 사용될 수 있다.

#### 1-1. 신경망의 발전
- 퍼셉트론(1950년대)
> 하나의 신경 세포와 같은 퍼셉트론을 만들어 냄 / 
> 하나의 입력에 대해 처리하고 결과를 출력

- 역전파(1982년대)
> Feedback을 할 수 있도록 해서 퍼셉트론의 결과가 틀린 경우 다시 퍼셉트론으로 입력 / 결과변수 Y가 틀린 경우 가중치를 조정하여 다시 퍼셉트론에 입력하면서 학습 수행

#### 1-2. 최신 신경망이 이슈되는 이유
- 컴퓨터 처리능력의 급증으로 1000억의 신경을 만들지 않아도 인간처럼 처리능력이 향상됨

#### 1-3. 신경망의 활용 분야
- 금융회사의 고객관리
- 주식 가격 시계열 예측
- 생산공정 관리
- 기업과 개인의 파산의 예측
- 의학적 진단
- 안면인식
- 자동주행

#### 1-4. 신경망의 장점과 단점

|장점|단점|
|----|----|
|- 예측 성능 우수|- 결과에 대한 해석 어려움|
|- 변수들 사이 복잡한 관계 우수하게 파악|- 복잡한 학습과정으로 모형 구축에 많은 시간 소요|
|- 데이터에 잡음이 많아도 좋은 성능을 냄|- 학습을 위해서는 많은 양의 데이터 필요|

### 2. 퍼셉트론
- 하나의 신경망에 해당되는 것으로 입력값의 데이터가 임계치를 넘으면 다른 퍼셉트론에 결과 전송
- 인간의 세포도 입력값을 받고 다른 세포에 전달할지를 결정하는 것과 유사
- 다수의 입력값과 가중치를 선형으로 결합하여 계산
- 활성화 함수에 따라 생성되는 출력값이 결정

- x0 ~ xn은 퍼셉트론 알고리즘의 입력값
- W0~Wn 은 x0 ~ xn 에 곱해지는 가중치 값이고 모든 가중치와 xn을 곱한 결과는 모두 더해서 하나의 결과값을 만듦. 신경망은 가중치 값을 사용해서 출력 조정.
- 활성함수는 최종 결과값을 다른 퍼셉트론에게 데이터로 전달할 것인지 않을 것인지 결정하는 것으로 로지스틱 회귀모델에서는 S Curve인 시그모이드 함수를 사용. 하지만 신경망에서는 다양한 함수 사용 가능!
- 학습이란 결과적으로 W0 ~ Wn을 찾아내는 과정

$$(f(3.1) > T(3)) = 1$$

#### ✅ 전방전파와 역전파
- 전방전파 : 학습데이터가 입력층으로 입력되면 은닉층의 활성화 함수에 의해 계산되어 최종 출력이 나오는 것
- 역전파 : 출력값을 다시 입력층으로 되돌리는 것으로 임의의 미분 가능한 비선형 활성화 함수

### 3. 활성화 함수(Activation Function)
- 어떤 신호를 입력 받아서 적절한 처리를 하고 출력하는 하뭇

- 활성화 함수를 사용해서 출력된 신호가 다음 단계에서 활성화 되는지를 결정

1,\;x>0
\end{cases}$$

#### 3-2. Sigmoid function
- 항상 0과 1 사이의 값만 가질 수 있는 비선형 함수
- Sigmoid 함수는 0과 1 사이의 값을 부드럽게 만들어 준다.
- 로지스틱 회귀분석에서 사용

#### 3-3. ReLU(Rectified Linear Unit) function
- 선형함수로 최근에 가장 많이 사용하고 있는 활성화 함수
- Sigmoid function의 Gradient Vanishing 문제를 해결하기 위해 ReLU를 사용
- ReLU 함수는 Deep Learning에서 많이 사용

##### ✅ Gradient Vanishing
- Sigmoid function은 0과 1 사이의 출력값을 가질 때 0에 매우 가까운 값을 가지게 되는 것 ex. 0.00001과 같은 출력
- Back-propagation 시에 각 Layer를 많이 통과하면 결국 0으로 수렴하는 문제

$$f = \begin{cases}
(x<0),\;f(x)=0\\
(x\geq0),\;f(x)=x
\end{cases}$$

### 5. Back-propagation(역전파 알고리즘)
- 신경망에서 학습이란 예측결과를 찾기 위해 연결강도인 임계치, 가중치를 찾는 것
- 비용함수, 손실함수는 에러를 최소화하기 위해 비용이 0에 가까운 시점을 찾는 것이다. 즉, 변동량을 계산하기 위해 미분을 사용해 비용이 작은 것을 탐색
- 역전파 알고리즘은 비용값이 0이 되도록 가중치를 조정하여 다시 입력값으로 전송

#### 5-1. 경사하강법(비용이 0에 가까운 시점을 찾는 것)

![경사하강법](https://blog.kakaocdn.net/dn/egkkO4/btqyiz3ybBg/Nr1Vb0pqAB2NkAhXIVFsXK/img.png)

출처 : https://it-license.tistory.com/19

- 비용함수를 사용해서 0에 가까워질 때까지 계산

$$error = \displaystyle\sum_{i=1}^{n} (y_i - \hat{y_i})^2$$

- 최소 비용을 찾으려고 탐색을 한 후에 최소비용을 설정하는 오류 발생 가능
- 이를 위해서 어느 시점까지 계속 탐색하는 모멘텀(Momentum)을 설정해야 함
- 즉, 모멘텀은 어느 정도 기존의 방향을 유지할 것인지 조정하는 것

![모멘텀 설정](https://ichi.pro/assets/images/max/724/0*d3qXU_U2gCUVgQ6F)

출처 : https://ichi.pro/ko/adam-adagrad-mich-rmsprop-ui-tansaeng-199220097497549

- Local Minimum을 선택하면 잘못된 최소비용
- 모멘텀을 설정하면 실제 최소비용 찾을 수 있을 것

### 7. 신경망의 문제점

#### 7-1. 대표적 단점
- 결과에 대한 해석이 어려운 Black box model
- 학습 완료 시점을 예측할 수 없음
- 학습이 완료될 때까지 많은 회수의 반복학습을 해야 함
- 추가 학습이 발생하면 전체적으로 재학습 해야 함

#### 7-2. 신경망의 포화상태와 과대적합
- 포화는 신경망에서 학습하면 할수록 활성화 함수에서 만들어질 수 없는 더 큰 결과값을 만들기 위해서 더 큰 가중치를 시도하게 되므로 신경망이 포화가 됨

##### ✅ 신경망 과대적합과 과소적합

|과소적합|- 분석모델이 너무 단순해서 데이터의 내재된 구조를 학습하지 못할 때 발생|
